```python
class InvertedIndex:
    def __init__(self):
        self.__content = {}
    
    def insert(self, word, doc):
        if word in self.__content:
            self.addOccurance(self.__content[word], doc)
        else:
            self.__content[word] = [1, {doc:  1}]
    
    def addOccurance(self, inverted_list, doc):
        print(inverted_list)
        print(doc)
        if doc in inverted_list[1]:
            print("Hits")
            inverted_list[1][doc] += 1
        else:
            print("no hit")
            inverted_list[1][doc] = 1
            inverted_list[0] += 1
            
    def __str__(self):
        return str(self.__content)
```

```python
import os

docs_path = '../Dataset/docs/'
docs_list = os.listdir(docs_path)

invertedIndex = InvertedIndex()

for doc_name in docs_list:
    file_path = os.path.join(docs_path, doc_name)
    doc = open(file_path)
    for word in doc:
        invertedIndex.insert(word[:-1], doc_name)
```
